# **Statstics_advance_part1**

1. What is a random variable in probability theory?
- A random variable is a variable that takes on numerical values based on the outcomes of a random experiment.

2. What are the types of random variables?
- There are two types:

  - Discrete random variables (finite/countable outcomes)

  - Continuous random variables (infinite/unmeasurable outcomes)

3. What is the difference between discrete and continuous distributions?

- Discrete: Values are countable (e.g., number of heads).

- Continuous: Values are measurable on a scale (e.g., height, weight).

4. What are probability distribution functions (PDF)?
- A PDF gives the probability that a random variable takes a particular value (for discrete) or lies within a range (for continuous).

5. How do cumulative distribution functions (CDF) differ from PDF?
- A CDF gives the probability that a random variable is less than or equal to a value, while a PDF gives the probability at a specific value or range.

6. What is a discrete uniform distribution?
- It's a distribution where all outcomes have equal probability, like rolling a fair die.

7. What are the key properties of a Bernoulli distribution?

- Only two outcomes: success (1) or failure (0)

- Single trial

- Probability of success = p, failure = 1 - p

8. What is the binomial distribution, and how is it used in probability?
- Describes the number of successes in n independent Bernoulli trials. Used in scenarios like "How many heads in 10 coin tosses?"

9. What is the Poisson distribution and where is it applied?
- Models the number of events occurring in a fixed interval (time, space) when events happen independently. Example: Calls per hour at a call center.

10. What is a continuous uniform distribution?
- Every value in a given range is equally likely, like picking a random number between 0 and 1.

11. What are the characteristics of a normal distribution?

- Bell-shaped and symmetric

- Mean = median = mode

- Defined by mean (μ) and standard deviation (σ)

12. What is the standard normal distribution, and why is it important?
- A normal distribution with mean = 0 and SD = 1. It simplifies calculations using Z-scores.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- CLT states that the sampling distribution of the sample mean approaches normality as the sample size increases, regardless of the population’s shape. It enables hypothesis testing.

14. How does the Central Limit Theorem relate to the normal distribution?
- CLT allows us to use the normal distribution to approximate the behavior of sample means, especially for large samples.

15. What is the application of Z statistics in hypothesis testing?
- Z-statistics are used to determine how far a sample statistic is from the population mean, helping to accept or reject a hypothesis.

16. How do you calculate a Z-score, and what does it represent?
- Z = (X - μ) / σ
- It tells how many standard deviations a value is from the mean.

17. What are point estimates and interval estimates in statistics?

- Point estimate: Single best guess (e.g., sample mean)

- Interval estimate: Range within which the true value lies, with confidence

18. What is the significance of confidence intervals in statistical analysis?
- They provide a range of likely values for a population parameter and show the precision of the estimate.

19. What is the relationship between a Z-score and a confidence interval?
- Z-scores are used to determine the margins of a confidence interval (e.g., Z=1.96 for 95% confidence).

20. How are Z-scores used to compare different distributions?
- By converting scores to Z-scores, you can compare values across different distributions with different means and variances.

21. What are the assumptions for applying the Central Limit Theorem?

- Random sampling

- Independent observations

- Large sample size (typically n ≥ 30)

22. What is the concept of expected value in a probability distribution?
- It’s the long-run average or mean value of a random variable over many trials.

23. How does a probability distribution relate to the expected outcome of a random variable?
- It provides the probabilities for each possible value, which are used to calculate the expected value (mean).

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

x = np.arange(1, 7)
pmf = randint.pmf(x, 1, 7)

plt.stem(x, pmf)
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("x")
plt.ylabel("PMF")
plt.grid()
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset

```python
from scipy.stats import norm
import numpy as np

sample = np.random.normal(100, 15, 30)
population_mean = 100
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-Statistic:", z)
print("P-Value:", p_value)
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 10, 50)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = 1.96  # 95% confidence
margin = z * (std / np.sqrt(n))

ci_lower = mean - margin
ci_upper = mean + margin

print("95% Confidence Interval:", (ci_lower, ci_upper))
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

21. Write a Python program to compare the standard deviations of two datasets and visualize the difference
